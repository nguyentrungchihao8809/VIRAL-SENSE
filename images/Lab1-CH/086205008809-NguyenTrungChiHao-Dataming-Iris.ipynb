```python
import pandas as pd
import matplotlib.pyplot as plt
import re 

# Đọc file dữ liệu
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, names=column_names)
print("Đọc dữ liệu thành công!")
```

```python
# 1. Hiển thị 5 dòng đầu tiên
print("5 dòng đầu tiên của tập dữ liệu:")
print(df.head())
```

```python
# 2. Cho biết kiểu dữ liệu các cột
print("Kiểu dữ liệu của các cột:")
print(df.dtypes)
```

```python
# 3. Hiển thị các giá trị thống kê
print("Bảng thống kê mô tả:")
print(df.describe())
```

```python
# 4. Trực quan hóa histogram cho các thuộc tính
df.hist(figsize=(10, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle('Histogram các thuộc tính của hoa Iris')
plt.show()
```
