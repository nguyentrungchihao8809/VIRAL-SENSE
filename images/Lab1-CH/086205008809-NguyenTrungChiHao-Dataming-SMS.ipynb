```python
import pandas as pd
import collections
import re

# Đọc file dữ liệu (File SMSSpamCollection cần đặt chung thư mục với file notebook này)
df = pd.read_csv('SMSSpamCollection', sep='\t', names=['label', 'message'], header=None)

# Kiểm tra dữ liệu
df.head()
```

```python
# 1. Đếm dòng và phân loại
total = len(df)
counts = df['label'].value_counts()

print(f"Tổng số dòng trong dữ liệu: {total}")
print(f"Số lượng tin nhắn Ham: {counts.get('ham', 0)}")
print(f"Số lượng tin nhắn Spam: {counts.get('spam', 0)}")
```

```python
# Gộp tất cả message thành một chuỗi lớn và chuyển về chữ thường
all_text = " ".join(df['message']).lower()

# Tách từ: Chỉ lấy các từ là chữ cái (bỏ qua số và ký tự đặc biệt)
words = re.findall(r'\b[a-z]+\b', all_text)
```

```python
# 2. Đếm số lượng từ
print(f"Tổng số từ trong dữ liệu: {len(words)}")
```

```python
# 3. Đếm số lần từ 'you' xuất hiện
you_count = words.count('you')
print(f"Số lần từ 'you' xuất hiện: {you_count}")
```

```python
# 4. Đếm số lượng từ khác nhau
unique_words = set(words)
print(f"Số lượng từ khác nhau: {len(unique_words)}")
```

```python
# 5. Liệt kê 10 từ khóa xuất hiện nhiều nhất
word_counts = collections.Counter(words)
top_10 = word_counts.most_common(10)

print("10 từ xuất hiện nhiều nhất là:")
for word, count in top_10:
    print(f"- '{word}': {count} lần")
```
