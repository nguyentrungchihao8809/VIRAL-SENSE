## 2. Kiểu dữ liệu cơ bản
- **Mô tả:** Các kiểu dữ liệu đơn giản nhất: String (chuỗi), Integer (số nguyên), Float (số thực), Bool (logic).

```python
s = "Python"    # String
i = 10          # Integer
f = 3.14        # Float
b = True        # Bool
print(f"Kiểu dữ liệu: {type(s)}, {type(i)}, {type(f)}, {type(b)}")
```

## 3. Kiểu dữ liệu nâng cao
- **List:** Danh sách có thứ tự, thay đổi được. Dùng lưu nhiều phần tử.
- **Tuple:** Danh sách có thứ tự, KHÔNG thay đổi được. Dùng lưu dữ liệu cố định.
- **Set:** Tập hợp các phần tử duy nhất (không trùng). Dùng để lọc trùng hoặc tính toán tập hợp.
- **Dictionary:** Cặp Key-Value. Dùng để tra cứu dữ liệu theo tên (như danh bạ).

```python
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_set = {1, 2, 2, 3} # Chỉ lưu {1, 2, 3}
my_dict = {"MSSV": "086205008809", "Ten": "Nguyen Trung Chi Hao"}
print(f"List: {my_list}, Set: {my_set}, Dict: {my_dict['Ten']}")
```

## 4. Cấu trúc lệnh
- **4.a. Lệnh điều kiện (if-else):** Dùng để phân nhánh chương trình dựa trên điều kiện.
- **4.b. Lệnh lặp (for/while):** Dùng để lặp đi lặp lại một hành động.

```python
# Lệnh điều kiện
diem = 8
if diem >= 5:
    print("Đạt")
else:
    print("Không đạt")

# Lệnh lặp
for i in range(3):
    print("Số thứ tự:", i)
```

## 5. Khai báo hàm
- **Mô tả:** Dùng `def` để gom nhóm các đoạn code dùng lại nhiều lần.

```python
def chao_hoi(ten):
    return "Chào " + ten
print(chao_hoi("Nguyen Trung Chi Hao"))
```

## 6. Đọc ghi file văn bản
- **Mô tả:** Dùng `open(filename, mode)` với mode 'w' (ghi) và 'r' (đọc).

```python
# Ghi file
with open("thongtin.txt", "w", encoding="utf-8") as f:
    f.write("MSSV: 086205008809 - Nguyen Trung Chi Hao")

# Đọc file
with open("thongtin.txt", "r", encoding="utf-8") as f:
    print(f.read())
```
